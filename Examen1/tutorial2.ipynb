```python
import pandas as pd
wines = pd.read_csv('winemag-data_first150k.csv', index_col=0)
wines.head()
```

```python
## Scatter plot
wines[wines['price'] < 100].sample(100).plot.scatter(x='price', y='points')
```

```python
wines[wines['price'] < 100].plot.scatter(x='price', y='points')
```

```python
## Hex plot
wines[wines['price'] < 100].plot.hexbin(x='price', y='points', gridsize=15)
```

## Exercises

```python
pokemon = pd.read_csv('Pokemon2.csv')
pokemon.head()
```

```python
pokemon.plot.scatter(x='Attack',y='Defense')
```

```python
pokemon.plot.hexbin(x='Attack',y='Defense',gridsize=15)
```

```python
pokemon_stats_legendary = pokemon.groupby(['Legendary', 'Generation']).mean()[['Attack', 'Defense']]
pokemon_stats_legendary.head(3)
```

```python
pokemon_stats_legendary.plot.bar(stacked=True)
```

```python
pokemon_stats_by_generation = pokemon.groupby('Generation').mean()[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
pokemon_stats_by_generation.head(3)
```

```python
pokemon_stats_by_generation.plot.line()
```
