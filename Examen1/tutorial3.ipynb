```python
import pandas as pd
wines = pd.read_csv('winemag-data_first150k.csv')
wines.head()
```

```python
import seaborn as sns
```

```python
## Seaborn countplot equivalent to a pandas bar chart
sns.countplot(wines['points'])
```

```python
## KDE Plot
sns.kdeplot(wines.query('price < 200').price)
```

```python
wines[wines['price'] < 200]['price'].value_counts().sort_index().plot.line()
```

```python
## Two dimentions kde
sns.kdeplot(wines[wines['price'] < 200].loc[:, ['price', 'points']].dropna().sample(5000))
```

```python
## Equivalent to pandas histogram
## Distplot
sns.distplot(wines['points'], bins=10, kde=False)
```

```python
## Jointplot
## Scatterplot copy:
sns.jointplot(x='price',y='points',data=wines[wines['price']<100])
```

```python
# Hex Plot copy:
sns.jointplot(x='price',y='points',data=wines[wines['price']<100], kind='hex', gridsize=20)
```

```python
## Box plot
df = wines[wines.variety.isin(wines.variety.value_counts().head(5).index)]
sns.boxplot(
    x='variety',
    y='points',
    data=df
)
```

```python
## Violin plot
sns.violinplot(
    x='variety',
    y='points',
    data=wines[wines['variety'].isin(wines['variety'].value_counts()[:5].index)]
)
```
